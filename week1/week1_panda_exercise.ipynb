```python
#Draw a histogram


#Step 1: Load the life expectancy data from the gapminder_lifeexpectancy.xlsx file into a pandas DataFrame.

import pandas as pd 

df = pd.read_excel("gapminder_lifeexpectancy.xlsx", index_col=0)
df
```

```python

#Step 2: Select a few columns with years. Note that the columns read from 
#Excel are integer values.

lifeexp = df[[1800, 1825, 1850, 1875, 1900, 1925, 1950, 1975, 2000, 2015]]
lifeexp
```

```python
#Step 3: Draw a histogram for 2015 with the default settings.

import matplotlib.pyplot as plt

lifeexp[2015].hist()
plt.savefig('histo.png')
```

```python
#Step 4: Try a few values for the number of bins. 
#Step 5: Modify the plot using some matplotlib functions.
#Step 7: Try some optional parameters for the hist() function, e.g.:


lifeexp[2015].hist(bins=25, facecolor='#00cccc', alpha=0.75,histtype='bar')
plt.title('Life_Expectancy')
plt.xlabel('Age')
plt.ylabel('Number_Of_Countries')
```

```python
#Step 6: Set the min/max x and y values with plt.axis.

lifeexp[2015].hist(bins=25, facecolor='#00cccc', alpha=0.75,histtype='bar')
plt.title('Life_Expectancy')
plt.xlabel('Age')
plt.ylabel('Number_of_Countries')
plt.axis([40, 90, 0, 25])
plt.show()
```

```python
#Step 8: Set the resolution of the image that is written.
plt.savefig('histo.png', dpi=150)
```
