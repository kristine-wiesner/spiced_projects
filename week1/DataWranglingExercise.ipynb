```python
#Combine two Dataframes
'''
 load the data in the file data/gapminder_lifeexpectancy.xlsx
 select a year from the DataFrame
 load the data in the file data/continents.csv
 merge both DataFrames
 remove all rows with empty values
 print the highest and lowest life expectancy for each continent
 plot the numbers
'''

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

life = pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col=0,)

life.index.name='country'

life
```

```python
life[1978]
```

```python
cont = pd.read_csv("continents.csv", index_col=1, delimiter=";")

#cont = cont.set_index('country', cont)

#cont = cont['country', 'continent']
cont
```

```python
lifecont = pd.merge(left=life, right=cont, how='inner', on='country')
lifecont
```

```python
lifecont.dropna()
```

```python
lifecont_asia = lifecont[lifecont['continent'] == 'Asia']
lifecont_asia.columns = lifecont_asia.columns.astype(str)
lifecont_asia.loc[:, '1990':'2000']
```
