```python
#Step 1 Load the file data/gapminder_total_fertility.csv into pandas.

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import imageio


from IPython.display import Image


fert = pd.read_csv('gapminder_total_fertility.csv', index_col=0)
fert

```

```python
#Step 2 Proceed in the same way with the file data/gapminder_lifeexpectancy.xlsx. Save it in a DataFrame with the name life. Use the function pandas.read_excel() to read in the excel file.

life = pd.read_excel('gapminder_lifeexpectancy.xlsx', index_col=0)
life
```

```python
#Step 3 Check whether both tables have the same size and shape.

print(fert.shape)
print(life.shape)
```

```python
#Step 4 Take a look at the columns of both tables:

#fert.columns
fert.columns = fert.columns.astype(int)
fert.columns
```

```python
#life.columns
life.columns = life.columns.astype(int)
life.columns
```

```python
#Step 5 Now, let’s check the row index of the table: 
#The name of the index should be called country. You can change it with:

fert.index
fert.index.name = 'country'
fert.index
```

```python
#Step 6  Now convert the table into long format. First, we have to move the row index into a column:

fert = fert.reset_index()

```

```python
# ...continue Step 6 ...Second, we use the convenient melt function for the conversion:

fert = fert.melt(id_vars='country', var_name='year', value_name='fertility_rate')
fert
```

```python
#Step 7  Repeat steps 5-6 for the life expectancy data, so that it’s also in long format.

life.index
life.index.name = 'country'
life = life.reset_index()
life = life.melt(id_vars='country', var_name='year', value_name='life_expectancy')
life
```

```python
#Step 8  To make things even more interesting, let’s also introduce population 
# data from the file data/gapminder_population.xlsx. Read the data in with pandas, 
# and get that DataFrame into the same format as the other two.

pop = pd.read_excel('gapminder_population.xlsx', index_col=0)
pop
```

```python
pop.index
pop.index.name = 'country'
pop = pop.reset_index()
pop = pop.melt(id_vars='country', var_name='year', value_name='population')
pop
```

```python
#Step 9  Now we can start glueing the DataFrames together. This works great with the merge function:

fert_pop = fert.merge(pop)
fert_pop
```

```python
fert_pop_life = fert_pop.merge(life)
fert_pop_life = fert_pop_life.dropna()
fert_pop_life
```

```python
#Step 10  With all the data merged into a single object, you can now select specific columns and plot them:


sns.set()

countries = ['Philippines','Indonesia', 'Vietnam', 'Thailand', 
     'Malaysia','Singapore','Myanmar', 'Brunei', 'Laos', 'Cambodia']

southeastasia = fert_pop_life.loc[
    (fert_pop_life['year'] >= 1950) & (fert_pop_life['year'] <= 2015)].loc[fert_pop_life['country'].isin(
    countries)]



sns.relplot(x='year', 
            y='life_expectancy', 
            hue='country',
            kind='line',
            data=southeastasia)
```

```python
sns.relplot(x='year', 
            y='fertility_rate', 
            hue='country',
            kind='line',
            data=southeastasia)
```

```python
sns.relplot(x='year', 
            y='population', 
            hue='country',
            kind='line',
            data=southeastasia)
```

```python
#fertility rate vs life expectancy

sns.relplot(x='life_expectancy', 
            y='fertility_rate', 
            hue='country',
            kind='line',
            data=southeastasia)
```

```python
#Step 11 To create an animated scatterplot showing the correlation between life expectancy and fertility,
#create one plot for each year between 1960 and 2015 (before that, the data contains too many gaps).
#Save each scatterplot to a separate file containing the year in the filename, e.g. lifeexp_1999.png .
#Hint
#Use the function plt.axis((xmin, xmax, ymin, ymax)) to fix the coordinates of the plot. 

def plot(df_subset):
    sns_plot = sns.scatterplot(x='life_expectancy', 
                               y='fertility_rate', 
                               hue='country',
                               data=df_subset, alpha=0.6)
    sns_plot.axis((0,100,0,10))
    plt.legend(loc='upper left')
    fig = sns_plot.get_figure()
    fig.savefig("lifeexp_"+str(year)+".png")
    fig.clf()
    
for year in range (1950,2015): 
    plot(southeastasia.loc[southeastasia['year'] == year].loc[southeastasia['country'].isin(countries)])
    
    

```

```python
#Step 12  Add another dimension to the visualization by utilizing the population data.

#Adjust the plot, so that the size of the symbols represent the population from the gapminder dataset.

def plot_pop(df_subset):
    sns_plot = sns.relplot(x='life_expectancy', y='fertility_rate', 
                           hue='country',
                           size='population',
                           sizes=(20, 200),
                           data=df_subset, alpha=0.6)
    sns_plot.set(ylim=(0,10))  
    sns_plot.set(xlim=(0,100))
    sns_plot.savefig("lifeexppop_"+str(year)+".png")
   
   
for year in range (1950,2015): 
    plot_pop(southeastasia.loc[southeastasia['year'] == year].loc[southeastasia['country'].isin(countries)])

    

```
