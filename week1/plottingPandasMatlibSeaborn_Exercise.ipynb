```python
#Plotting with Pandas
# Create a DataFrame with a x and y column
# Call df.plot()
# Fine-tune the plot afterwards (see section matplotlib)

import pandas as pd

x = pd.Series(range(-10,11))
y = x**2

df = pd.DataFrame({
    'x': x,
    'y': y
})

df.plot(x ='x',
        y='y',
        kind = 'line', 
        title = 'Polynomial y=x**2', 
        grid = True, legend = True)
```

```python
#Plotting with Matplotlib

''' set the size of your figure with plt.figure(figsize=(<width>,<height>))
    call plt.plot(x, y) or plt.scatter(x, y)
    Modify title, x labels or limits
    add a legend
    save the plot
'''

import pandas as pd
import matplotlib.pyplot as plt

x = pd.Series(range(-10,11))
y = x**2

df = pd.DataFrame({
    'x': x,
    'y': y
})
plt.figure(figsize=(10,5))
plt.plot(y)
plt.title('Polynomial y**2')    
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend('y')
plt.savefig('polynomial_matplot.png')
```

```python
#Plotting with Seaborn
'''
    Create a DataFrame with a x and y column
    Use one of the high-level functions (e.g. sns.lineplot(), see here: https://seaborn.pydata.org/api.html)
    Fine tune the plot afterwards (see section matplotlib)
'''


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-white')

x = pd.Series(range(-10,11))
y = x**2

df = pd.DataFrame({
    'x': x,
    'y': y
})


sns.lineplot(
    data=df, 
    x="x", 
    y="y"
)

sns.despine()
plt.title('Polynomial y**2')    
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend('y')
```
